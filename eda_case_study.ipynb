```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
```

```python
application = pd.read_csv("application_data.csv\\application_data.csv")
pre_application = pd.read_csv("previous_application.csv\previous_application.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='NAME_EDUCATION_TYPE', y='TARGET', data=application)
plt.title('Default Rate by Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(application.loc[application['TARGET'] == 0, 'DAYS_BIRTH'] / -365, label='Target 0', shade=True)
sns.kdeplot(application.loc[application['TARGET'] == 1, 'DAYS_BIRTH'] / -365, label='Target 1', shade=True)
plt.xlabel('Age (years)')
plt.title('Age Distribution by Target')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(application.loc[application['TARGET'] == 0, 'EXT_SOURCE_3'], label='Target 0', shade=True)
sns.kdeplot(application.loc[application['TARGET'] == 1, 'EXT_SOURCE_3'], label='Target 1', shade=True)
plt.title('EXT_SOURCE_3', fontsize=15)
plt.xlabel('Credit')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
target_map = application[['SK_ID_CURR', 'TARGET']]


prev_with_target = pre_application.merge(target_map, on='SK_ID_CURR', how='left')
plt.figure(figsize=(10, 6))

res = prev_with_target.groupby('NAME_CONTRACT_STATUS')['TARGET'].mean().sort_values(ascending=False)

sns.barplot(x=res.index, y=res.values, palette='magma')
plt.axhline(target_map['TARGET'].mean(), color='blue', linestyle='--', label='Overall Avg Default')
plt.title('Current Default Rate by Previous Application Status', fontsize=14)
plt.ylabel('Proportion of Current Defaults')
plt.legend()
plt.show()
```
